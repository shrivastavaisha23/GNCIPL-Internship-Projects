# **Summary**

This project analyzes cyberattack trends by sector using the "cyber_incidents.csv" dataset. *    **DATA-**  "cyber_incidents.csv"  (Kaggle)
*  **OBJECTIVE-** The primary objective of this Exploratory Data Analysis is to investigate and understand the trends in cyberattacks across different industry sectors.


*   **TOOLS & LIBRARIES-**  Python, pandas, matplotlib and seaborn.








Imports and Load Data:

```python
import pandas as pd

# Load your dataset
df = pd.read_csv("../data/raw/cyber_incidents.csv")

# See first 5 rows
df.head()

```

Check dataset shape and columns

```python
# Check dataset shape and columns
df.shape, df.columns
```

Uses the matplotlib.pyplot and seaborn libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count attacks per sector
sector_counts = df['sector'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=sector_counts.index, y=sector_counts.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title("Frequency of Cyberattacks by Sector")
plt.xlabel("Sector")
plt.ylabel("Number of Incidents")
plt.show()
```

import matplotlib.pyplot as plt: Imports the matplotlib.pyplot library, commonly used for creating plots and visualizations, and assigns it the alias plt. import seaborn as sns: Imports the seaborn library, which is built on top of Matplotlib and provides a high-level interface for drawing attractive statistical graphics, and assigns it the alias sns. sector_counts = df['sector'].value_counts(): Calculates the number of occurrences for each unique value in the 'sector' column of the DataFrame df and stores the results in the sector_counts variable. plt.figure(figsize=(10,6)): Creates a new figure for plotting with a specified size of 10 inches in width and 6 inches in height. sns.barplot(x=sector_counts.index, y=sector_counts.values, palette="viridis"): Generates a bar plot using seaborn. The x-axis represents the unique sectors (from sector_counts.index), the y-axis represents the count of incidents for each sector (from sector_counts.values), and the bars are colored using the "viridis" palette. plt.xticks(rotation=45, ha='right'): Rotates the x-axis labels by 45 degrees to prevent overlap and aligns them to the right. plt.title("Frequency of Cyberattacks by Sector"): Sets the title of the plot. plt.xlabel("Sector"): Sets the label for the x-axis. plt.ylabel("Number of Incidents"): Sets the label for the y-axis. plt.show(): Displays the generated plot

 Count incidents per year

```python

yearly_counts = df['year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker="o")
plt.title("Cyberattacks Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Incidents")
plt.show()
```


yearly_counts = df['year'].value_counts().sort_index(): This calculates the number of incidents for each year by counting the occurrences of each year in the 'year' column and then sorts the results by year.
plt.figure(figsize=(10,6)): This creates a new figure for the plot with a specified size.
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker="o"): This generates the line plot using seaborn.
x=yearly_counts.index: Sets the x-axis values to the years (the index of the yearly_counts Series).
y=yearly_counts.values: Sets the y-axis values to the incident counts for each year (the values of the yearly_counts Series).
marker="o": Adds circular markers at each data point on the line.
plt.title("Cyberattacks Over Time"), plt.xlabel("Year"), plt.ylabel("Number of Incidents"): These lines set the title of the plot and the labels for the x and y axes.
plt.show(): This displays the generated plot.

Count attack vectors

```python

attack_counts = df['attack_vector'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=attack_counts.index, y=attack_counts.values, palette="mako")
plt.xticks(rotation=45, ha='right')
plt.title("Top Attack Vectors")
plt.xlabel("Attack Vector")
plt.ylabel("Number of Incidents")
plt.show()
```

 Heatmap of Cyberattacks by Sector & Severity


```python
plt.figure(figsize=(8,5))
heatmap_data = pd.crosstab(df["sector"], df["severity"])
sns.heatmap(heatmap_data, annot=True, cmap="Reds", fmt="d")
plt.title("Heatmap of Cyberattacks by Sector & Severity")
plt.show()
```

plt.figure(figsize=(8,5)): This line creates a new figure for the plot with a specified size of 8 inches in width and 5 inches in height.
heatmap_data = pd.crosstab(df["sector"], df["severity"]): This is a crucial step where pd.crosstab is used to create a cross-tabulation (a frequency table) of the 'sector' and 'severity' columns from your DataFrame df. This table shows how many incidents fall into each combination of sector and severity level.
sns.heatmap(heatmap_data, annot=True, cmap="Reds", fmt="d"): This line generates the heatmap using seaborn.
heatmap_data: Provides the data for the heatmap (the cross-tabulation table).
annot=True: Displays the numerical values (counts) on each cell of the heatmap.
cmap="Reds": Sets the color scheme for the heatmap to a gradient of reds, where darker shades indicate higher counts.
fmt="d": Formats the annotations as integers.
plt.title("Heatmap of Cyberattacks by Sector & Severity"): Sets the title of the heatmap.
plt.show(): Displays the generated heatmap.
This heatmap allows you to quickly see which sectors experience the most severe cyberattacks and the distribution of severity levels within each sector.



 Cyberattacks Distribution By Sector

```python
plt.figure(figsize=(7,7))
df["sector"].value_counts().plot.pie(autopct="%1.1f%%", startangle=140, shadow=True)
plt.title("Cyberattacks Distribution by Sector")
plt.ylabel("")
plt.show()
```

plt.figure(figsize=(7,7)): This line creates a new figure for the plot with a specified size of 7 inches in width and 7 inches in height.
df["sector"].value_counts().plot.pie(...): This is the core of the pie chart generation.
df["sector"].value_counts(): This calculates the number of occurrences for each unique value in the 'sector' column, giving you the count of incidents per sector.
.plot.pie(...): This is a pandas plotting function that generates a pie chart directly from the results of value_counts().
autopct="%1.1f%%": This formats the percentages displayed on each slice of the pie chart to one decimal place.
startangle=140: This rotates the starting point of the first slice by 140 degrees.
shadow=True: This adds a shadow effect to the pie chart.
plt.title("Cyberattacks Distribution by Sector"): Sets the title of the pie chart.
plt.ylabel(""): This removes the default y-axis label that might appear.
plt.show(): Displays the generated pie chart.
This pie chart provides a visual representation of the proportion of cyberattacks that occur in each sector, making it easy to see which sectors are most frequently targeted.



Total Records Affected BY Sector

```python
plt.figure(figsize=(8,5))
records_by_sector = df.groupby("sector")["records_affected"].sum().sort_values(ascending=False)
records_by_sector.plot(kind="bar")
plt.title("Total Records Affected by Sector")
plt.xlabel("Sector")
plt.ylabel("Records Affected")
plt.xticks(rotation=45)
plt.show()
```

plt.figure(figsize=(8,5)): This creates a new figure for the plot with a specified size of 8 inches in width and 5 inches in height.
records_by_sector = df.groupby("sector")["records_affected"].sum().sort_values(ascending=False): This line calculates the total number of records affected for each sector.
df.groupby("sector"): Groups the DataFrame by the 'sector' column.
["records_affected"].sum(): Calculates the sum of the 'records_affected' column for each sector group.
.sort_values(ascending=False): Sorts the results in descending order.
records_by_sector.plot(kind="bar"): This generates a bar plot from the calculated sums, where the x-axis represents the sectors and the y-axis represents the total records affected.
plt.title("Total Records Affected by Sector"): Sets the title of the plot.
plt.xlabel("Sector"): Sets the label for the x-axis.
plt.ylabel("Records Affected"): Sets the label for the y-axis.
plt.xticks(rotation=45): Rotates the x-axis labels by 45 degrees to improve readability.
plt.show(): Displays the generated plot.
This plot helps visualize which sectors experienced the highest impact in terms of the number of affected records.
