5-**Visualize Top 10 States by Cyber Crime (Pie Chart and Bar Chart)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\ishas\Downloads\archive (1)\cyber-crime.csv")

# Remove summary rows
df = df[~df['State/UT'].str.contains("TOTAL", case=False)]

# Group and sort
data = df.groupby('State/UT')['Total'].sum().sort_values(ascending=False)

# Take Top 10 states
top10 = data.head(10)

# Create figure with 2 plots
plt.figure(figsize=(12,6))

# Pie Chart (Top 10 states)
plt.subplot(1,2,1)
plt.pie(top10, labels=top10.index, autopct='%1.1f%%', startangle=90)
plt.title("Top 10 States by Cyber Crime (Pie Chart)")

# Bar Chart (Top 10 states)
plt.subplot(1,2,2)
plt.bar(top10.index, top10.values, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Total Cyber Crimes")
plt.title("Top 10 States by Cyber Crime (Bar Chart)")

plt.tight_layout()
plt.show()
```

6-**Overall Cyber Crime Trend in India (Line Plot)**

```python
import matplotlib.pyplot as plt

# Drop non-year columns and keep only numeric year columns
yearly = df.drop(columns=['id','Total','Total-Scale'], errors='ignore')
yearly = yearly.set_index('State/UT')

# Convert all columns to numeric (ignore errors for non-numeric)
yearly = yearly.apply(pd.to_numeric, errors='coerce')

# India-wide yearly total
india_trend = yearly.sum(axis=0)

# Plot
india_trend.plot(kind='line', figsize=(10,5), marker='o')
plt.title("Cyber Crime Trend in India (2002â€“2021)")
plt.ylabel("Total Cases")
plt.xlabel("Year")
plt.grid(True)
plt.show()
```

7-**State-wise Cyber Crime Trends (Line Plot)**

```python
# Pick top 5 states by total crimes
top_states = yearly.sum(axis=1).sort_values(ascending=False).head(5).index

# Transpose for plotting
yearly.T[top_states].plot(figsize=(12,6), marker='o')
plt.title("Cyber Crime Trend (Top 5 States, 2002â€“2021)")
plt.ylabel("Total Cases")
plt.xlabel("Year")
plt.legend(title="State/UT")
plt.grid(True)
plt.show()
```

8-**Heatmap of Cyber Crimes by State & Year**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop extra columns, keep only years
heatmap_data = df.drop(columns=['id','Total','Total-Scale'], errors='ignore')
heatmap_data = heatmap_data.set_index('State/UT')

# Convert everything to numeric (ignore errors for non-numeric)
heatmap_data = heatmap_data.apply(pd.to_numeric, errors='coerce')

# Remove the TOTAL rows to avoid duplication
heatmap_data = heatmap_data.drop(index=['TOTAL (ALL INDIA)', 'TOTAL (STATES)'], errors='ignore')

# Plot heatmap
plt.figure(figsize=(16,8))
sns.heatmap(heatmap_data, cmap="Reds", linewidths=0.5, annot=False)
plt.title("Heatmap of Cyber Crimes by State & Year (2002â€“2021)")
plt.xlabel("Year")
plt.ylabel("State/UT")
plt.show()
```

9-**Scatter Plot of Cyber Crimes in India (2002â€“2021)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\ishas\Downloads\archive (1)\cyber-crime.csv")

# Drop extra columns and set index
yearly_data = df.drop(columns=['id', 'Total-Scale'], errors='ignore')
yearly_data = yearly_data.set_index('State/UT')

# Remove 'Total' row if exists
yearly_data = yearly_data.drop('Total', axis=0, errors='ignore')

# Keep only year columns
year_cols = [col for col in yearly_data.columns if col.isdigit()]

# Convert year columns to numeric (important fix ðŸš€)
yearly_data[year_cols] = yearly_data[year_cols].apply(pd.to_numeric, errors='coerce')

# Get India totals (sum of all states)
totals = yearly_data[year_cols].sum()

# Convert index (years) to integers
totals.index = totals.index.astype(int)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(totals.index, totals.values, color='red')
plt.plot(totals.index, totals.values, linestyle='--', alpha=0.6)  # optional trend line
plt.title("Scatter Plot of Cyber Crimes in India (2002â€“2021)")
plt.xlabel("Year")
plt.ylabel("Number of Cases")
plt.grid(True)
plt.show()
```

12-**Scatter/Strip Plot of Cyber Crimes per State over Years**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Melt the dataset into long format
year_cols = [str(y) for y in range(2002, 2022)]
df_melt = df.melt(id_vars="State/UT", value_vars=year_cols,
                  var_name="Year", value_name="Crimes")

# Replace '--' with NaN and convert to numeric
df_melt["Crimes"] = pd.to_numeric(df_melt["Crimes"].replace("--", np.nan))

plt.figure(figsize=(14,7))

# Strip plot (dots) grouped by Year
sns.stripplot(x="Year", y="Crimes", data=df_melt, jitter=True, hue="State/UT", dodge=False, alpha=0.7)

plt.xticks(rotation=45)
plt.title("Scatter/Strip Plot of Cyber Crimes per State over Years")
plt.legend([],[], frameon=False)  # remove legend (too many states), or keep if you want
plt.show()
```
